### Load Dataset

### pivot table

```python
pd.pivot_table(order, index = "state", values = "price") #평균 (mean 이 defult)
```

```python
pd.pivot_table(order, index = "state", values = "price", aggfunc = ["mean","sum"]) #합계
```

```python
pd.pivot_table(order, index = "address", values = "price") #NaN은 통계 연산에서 배재된다
```

```python
import numpy as np

#aggfunc에 method 쓸 수 있음
pd.pivot_table(order, index = "state", values="price", aggfunc = [np.mean,np.sum])
```

```python
pd.pivot_table(order, index = ["date","state"], values="price", aggfunc = [np.mean,np.sum])
```

### Group by

```python
order.groupby("state")["price"].mean()
```

```python
order.groupby("state")["price"].sum()
```

```python
order.groupby(["date","state"])["price"].mean()
```

### apply

```python
def is_vip(price):
    if price >= 500:
        return "VIP"
    else: return "Non-VIP"
order["price"].apply(is_vip)
```

```python
def is_vip(row):
    if row["state"] == "confirmed" and row["price"] >=500 : 
        return "VIP"
    else : return "Non-VIP"

order.apply(is_vip, axis = "columns") #칼럼 여러개 넘기기
```

### 여러개 데이터 하나로 합치기
