```python
import pandas as pd
```

### Load Dataset


```python
url = "https://bit.ly/dsa-01-order"
order = pd.read_csv(url, index_col = "id")

# index 지정하기
# order = order.set_index("id", inplace = True)
order
```

### 정렬하기

```python
#index로 정렬
order.sort_index()
# order.sort_index(ascending=False)
```

```python
order.sort_index(axis = 1) #axis = 0 (index)은 index,1은 칼럼 행 의미
order.sort_index(axis = "columns")
```

```python
order.sort_index(axis = 1,ascending = False) #axis = 0 은 index,1은 칼럼 행 의미
```

```python
# 칼럼으로 정렬
order.sort_values(by = "price")
# order.sort_values(by = "price", ascending = False)
```

```python
#두개 기준 정렬
order.sort_values(by = ["date","price"])
```

```python
order.sort_values(by = ["date","price"],ascending = False)
```

```python
order.sort_values(by = ["date","price"],ascending = [True,False]) #date는 작은 순-> 같을 경우 price 큰순
```

```python
#Series 버전
order["price"].sort_values(ascending = False)
```

```python
#method chaning : method 중첩 (리턴 타입이 같은 것끼리 메소드 중첩해서 쓸수 있음)
```

```python
order.sort_values(by = "price", ascending = False).head() #매출 최상위 5개 
```

```python
order["price"].sort_values(ascending = False).head()
```
