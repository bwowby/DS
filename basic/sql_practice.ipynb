# Load Dataset

```python
connection = sqlite3.connect("e-commerce.sqlite")
connection
```

```python
query = "SELECT * FROM orders "
order = pd.read_sql(query,connection, index_col="id")
order
```

```python
query = "SELECT id,date,price FROM orders limit 3" 
order = pd.read_sql(query,connection, index_col="id")
order
```

```python
query = "SELECT * FROM orders order by date desc"
order = pd.read_sql(query,connection, index_col="id")
order
```

```python
query = """
    SELECT * 
    FROM orders 
    order by date desc
    limit 3
"""
order = pd.read_sql(query,connection, index_col="id")
order
```

```python
query = """
    SELECT * FROM orders
    where price between 400 and 500
"""
order = pd.read_sql(query,connection, index_col="id")
order
```

```python
query = """
    SELECT * FROM orders
    where state <> "confirmed"
    limit 3
"""
order = pd.read_sql(query,connection, index_col="id")
order
```

```python
query = """
    SELECT * FROM orders
    where price >= 500 and state <> "confirmed"
"""
order = pd.read_sql(query,connection, index_col="id")
order
```
