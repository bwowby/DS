# Precision Recall
T/F : 예측 맞/틀 P/N : 예측 값 1/0   

* **정확도** : Accuracy   
  *TP / TP+FP+FN+TN*

* **정밀도** : Precision   
  *TP / TP+FP*   
병이라고 판정한 사람 중 병 걸린 사람일 확률(모델 관점)   

* **재현율** : Recall   
  *TP / TP+FN*   
병 걸린 사람 중 병이라고 판정할 확률(실제 값 관점)   
민감도(Sensitivity)라고도 하며 ROC 커브에서 TPR   

* **민감도** : Sensitivity   
  *TN/FP+TN*   
  TNR,1-TNR인 FPR과 TPR은 비례관계 

* **F1** : 정밀도와 재현율의 조화 평균    
정밀도,재현율 두 값을 동시에 고려한 더 제대로 된 평가를 위한 하나의 대표값 (두 값이 모두 높아야 좋은 모델)   
데이터 label이 불균형 구조일 때 모델 성능을 정확하게 평가 가능. 조화평균이기에 더 큰 값을 가지는 지표의 bias를 줄여준다.

* 잘못 판정했을 경우: FN(제 1종오류), FP(제 2종 오류)

```
import numpy as np
import matplotlib.pyplot as
import sklearn.metrics as metrics
```

```
y = np.array([1,1,1,1,0,0]) # Actual Value
p = np.array([1,1,0,0,0,0]) # Predicted Value
```

```
#정확도 : TP+TN/TP+FP+FN+TN
accuracy = np.mean(np.equal(y,p))

# TP
right = np.sum(y*p ==1) 

#정밀도 : TP/TP+FP
precision = right/np.sum(p)

#재현율 : TP/TP+FN
recall = right/np.sum(y)

f1 = 2*precision*recall/(precision+recall)
```

```
print('accauray : ', accuracy)
print('precision : ',precision)
print('recall : ',recall )
print('f1 :', f1 )
```

```
#sklearn에 다 있음
print('accauray : ', metrics.accuracy_score(y,p))
print('precision : ', metrics.precision_score(y,p))
print('recall : ', metrics.recall_score(y,p))
print('f1 :',  metrics.f1_score(y,p))
```

```
print(metrics.classification_report(y,p))
```

```
print(metrics.confusion_matrix(y,p))
```
